```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax = plt.axes()
sns.heatmap(booksDF.isna().transpose(), cbar= False, annot=True, ax=ax)

plt.xlabel("Number of missing values in columns ")
plt.ylabel("Number of missing values in rows")

plt.show()
```

```python
import numpy as np
booksDF["missing_description"] = np.where(booksDF["description"].isna(), 1, 0)
booksDF["age_of_book"] = 2026 - booksDF["published_year"]
```

```python
columns_of_interest = ["num_pages", "age_of_book", "missing_description", "average_rating"]

correlation_matrix = booksDF[columns_of_interest].corr(method="spearman")
sns.set_theme(style="white")
plt.figure(figsize=(8,6))
heatmap = sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar_kws={"label": "Spearman correlation"})
heatmap.set_title("Correlation Heatmap")
plt.show()
```

```python
book_missing = booksDF[(booksDF["num_pages"].isna()) |
        (booksDF["description"].isna()) |
        (booksDF["published_year"].isna()) |
        (booksDF["average_rating"].isna()) ]
```

```python
book_missing = booksDF[~(booksDF["num_pages"].isna()) &
                       ~(booksDF["description"].isna()) &
                       ~(booksDF["published_year"].isna()) &
                       ~(booksDF["average_rating"].isna()) ]
book_missing
```

```python
book_missing["categories"].value_counts().reset_index().sort_values("count", ascending=False)
```

```python
book_missing["words_in_description"] = book_missing["description"].str.split().str.len()
book_missing
```

```python
book_missing.loc[book_missing["words_in_description"].between(10,14), "description"]
```

```python
book_missing.loc[book_missing["words_in_description"].between(15,24), "description"]
```

```python
book_missing.loc[book_missing["words_in_description"].between(25,34), "description"]
```

```python
book_missing_25_words = book_missing[book_missing["words_in_description"] >= 25]
```

```python
book_missing_25_words
```

```python
combined_text = book_missing_25_words["title"].astype(str) + ": " + book_missing_25_words["subtitle"].astype(str)
book_missing_25_words["title_and_subtitle"] = np.where(
    book_missing_25_words["subtitle"].isna(),
    book_missing_25_words["title"],
    combined_text
)
```

```python
book_missing_25_words
```

```python
book_missing_25_words["tagged_description"] = book_missing_25_words[["isbn13", "description"]].astype(str).agg(" ".join, axis=1)
```

```python
book_missing_25_words
```

```python
(
    book_missing_25_words.drop(["subtitle", "missing_description", "age_of_book","words_in_description"], axis=1).to_csv("books_cleaned.csv", index = False)
)
```
